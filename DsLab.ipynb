```python
import numpy as np

# --- Basic Arithmetic Operations ---
array1 = np.array([10, 20, 30])
array2 = np.array([1, 2, 3])

# Addition
print("Addition:", array1 + array2)

# Subtraction
print("Subtraction:", array1 - array2)

# Multiplication (element-wise)
print("Element-wise Multiplication:", array1 * array2)

# Division (element-wise)
print("Element-wise Division:", array1 / array2)

# Exponentiation (element-wise)
print("Element-wise Exponentiation:", array1 ** 2)

# --- Aggregate Functions ---
array_agg = np.array([1, 2, 3, 4, 5, 6])

# Sum
print("\nSum of array:", np.sum(array_agg))

# Mean
print("Mean of array:", np.mean(array_agg))

# Median
print("Median of array:", np.median(array_agg))

# Standard Deviation
print("Standard Deviation of array:", np.std(array_agg))

# Minimum
print("Minimum value:", np.min(array_agg))

# Maximum
print("Maximum value:", np.max(array_agg))

# --- Array Manipulation ---
array_manipulate = np.array([[1, 2], [3, 4]])

# Transpose
print("\nTranspose of array:\n", array_manipulate.T)

# Reshape
array_reshape = np.array([1, 2, 3, 4, 5, 6])
print("Reshaped array (2x3):\n", array_reshape.reshape(2, 3))

# Concatenate
array_concat1 = np.array([[1, 2], [3, 4]])
array_concat2 = np.array([[5, 6]])
print("Concatenated array (along axis 0):\n", np.concatenate((array_concat1, array_concat2), axis=0))

# --- Indexing and Slicing ---
array_index = np.array([10, 20, 30, 40, 50])

# Accessing a single element
print("\nElement at index 2:", array_index[2])

# Slicing
print("Elements from index 1 to 3:", array_index[1:4]) # Note: end index is exclusive

# --- Broadcasting ---
# When performing operations on arrays of different shapes, NumPy may "broadcast"
# the smaller array across the larger array so that they have compatible shapes.
array_broadcast = np.array([1, 2, 3])
scalar = 5
print("\nBroadcasting (array + scalar):", array_broadcast + scalar)

matrix_broadcast = np.array([[1, 2, 3], [4, 5, 6]])
row_vector = np.array([10, 20, 30])
print("Broadcasting (matrix + row vector):\n", matrix_broadcast + row_vector)

# --- Linear Algebra (basic example) ---
matrix_a = np.array([[1, 2], [3, 4]])
matrix_b = np.array([[5, 6], [7, 8]])

# Matrix multiplication
print("\nMatrix multiplication:\n", np.dot(matrix_a, matrix_b))
# Or using the @ operator (Python 3.5+)
print("Matrix multiplication (using @):\n", matrix_a @ matrix_b)
```

# Task
Analyze sales data from a NumPy array to calculate total sales revenue per product, average sales revenue per category, find the top 5 products by revenue, and determine the correlation between sales and price. The dataset is a NumPy array with columns for product_id, sales, price, and category.

## Create the dataset

### Subtask:
Generate the NumPy array containing the product data as specified.


**Reasoning**:
The subtask requires creating a NumPy array with mixed data types. A structured array is the most suitable way to achieve this in NumPy. The code will define the structure of the array and then populate it with sample data as instructed.



```python
import numpy as np

product_id = np.arange(1, 101)
sales = np.random.randint(1, 100, size=100)
price = np.random.uniform(10, 100, size=100)
category = np.random.choice(['electronics', 'clothing', 'home goods'], size=100)

# Now we need to combine these into a single NumPy array as described in the task.
# Since the data types are different, a structured array is suitable.
data_type = np.dtype([('product_id', int), ('sales', int), ('price', float), ('category', 'U20')])

# Create the structured array
sales_data = np.empty(100, dtype=data_type)
sales_data['product_id'] = product_id
sales_data['sales'] = sales
sales_data['price'] = price
sales_data['category'] = category

print("Sales Data Array:")
print(sales_data) # Print the first 5 rows to check
```

## Calculate total sales revenue per product

### Subtask:
Compute the element-wise product of the 'sales' and 'price' arrays.


**Reasoning**:
Extract the 'sales' and 'price' columns from the `sales_data` array and compute their element-wise product to get the total sales revenue for each product.



```python
# Extract the 'sales' and 'price' columns
sales = sales_data['sales']
price = sales_data['price']
# Compute the element-wise product to get total revenue
total_revenue_per_product = sales * price
# Display the calculated total revenue
print("Total Sales Revenue per Product:", total_revenue_per_product)
```

## Calculate average sales revenue per category

### Subtask:
Determine the average sales revenue for each product category. This step will require handling the categorical data with NumPy which can be a bit tricky, so we might need to use Pandas for easier grouping.


**Reasoning**:
Convert the NumPy array to a pandas DataFrame, calculate the revenue for each product, group by category, and calculate the mean revenue for each category.



```python
# Convert the NumPy array to a pandas DataFrame
df_sales = pd.DataFrame(sales_data)
# Calculate total revenue per product
df_sales['revenue'] = df_sales['sales'] * df_sales['price']
# Group by category and calculate the mean revenue
average_revenue_per_category = df_sales.groupby('category')['revenue'].mean()
# Print the result
print("Average Sales Revenue per Category:\n", average_revenue_per_category)
```

## Find top 5 products

### Subtask:
Identify the indices of the top 5 products based on their total sales revenue and extract their product IDs.


**Reasoning**:
Use argsort to get the indices of the top 5 products by revenue and extract their product IDs.



```python
# Get the indices that would sort the total_revenue_per_product array in ascending order
sorted_indices = np.argsort(total_revenue_per_product)
# Reverse the order to get indices in descending order of revenue
top_indices_desc = sorted_indices[::-1]
# Get the indices of the top 5 products
top_5_indices = top_indices_desc[:5]
# Extract the product IDs of the top 5 products
top_5_products_by_id = sales_data['product_id'][top_5_indices]
# Print the top 5 product IDs
print("Top 5 Product IDs by Revenue:", top_5_products_by_id)
```

## Calculate correlation

### Subtask:
Compute the correlation coefficient between the 'sales' and 'price' arrays.


**Reasoning**:
Compute the correlation coefficient between the 'sales' and 'price' arrays using np.corrcoef and print the result.



```python
# Compute the correlation matrix between sales and price
correlation_matrix = np.corrcoef(sales, price)
# Extract the correlation coefficient between sales and price (off-diagonal element)
correlation_coefficient = correlation_matrix[0, 1]
# Print the correlation coefficient
print(f"Correlation coefficient between sales and price: {correlation_coefficient}")
```

## Present results

### Subtask:
Display the calculated total sales revenue, average sales revenue per category, top 5 products, and the correlation coefficient.


**Reasoning**:
Display the calculated results as requested in the instructions.



```python
print("Total Sales Revenue per Product:", total_revenue_per_product)
print("Average Sales Revenue per Category:\n", average_revenue_per_category)
print("Top 5 Product IDs by Revenue:", top_5_products_by_id)
print(f"Correlation coefficient between sales and price: {correlation_coefficient}")
```
